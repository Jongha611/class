#### 더보기 버튼 누르기
- 한솥도시락 페이지는 더보기 버튼을 모두 누르지 않으면 모든 메뉴가 보이지 않는다.
- 더보기 버튼의 위치를 찾고
- click() 메소드를 사용하여 더보기버튼을 끝까지 눌러주자

```python
btn_more = driver.find_elements(By.CLASS_NAME, 'c_05')
btn_more = btn_more[0]
btn_more.click()
```

```python
# 더보기가 몇 번 눌러져야 모든 데이터가 보일지 알 수 없기 때문에
# => try/except 사용
try:
    # 에러가 나기 전까지 실행해야 할 실행문장 작성
    for i in range(300):
        btn_more = driver.find_elements(By.CLASS_NAME, 'c_05')
        btn_more = btn_more[0]
        btn_more.click()        
except:
    # try문을 실행하다가 오류가 발생했다면, except문 실행
    print('더보기 버튼 전부 클릭 완료')
```

```python
# 로드하고 있는 데이터의 형태 => 이미지 + 텍스트
# 이미지 데이터나 영상 데이터, 애니메이션 데이터 => 로드되는데 많은 메모리를 소모한다
# 용량이 큰 데이터들을 한 번에 로드할 경우 페이지에 무리가 살 수 있다
# 그리고 페이지에서 사람이 수행하는 작업이 아니라는 것을 눈치 챌 가능성이 있다
import time # time.sleep(몇 초?) => 몇 초만큼 기다렸다가 실행
```

```python
try:
    # 에러가 나기 전까지 실행해야 할 실행문장 작성
    for i in range(300):
        time.sleep(2) # 2초의 쉬는 시간이 부여됨 (페이지가 무거울 때 안정성 증가)
        btn_more = driver.find_elements(By.CLASS_NAME, 'c_05')
        btn_more = btn_more[0]
        btn_more.click()        
except:
    # try문을 실행하다가 오류가 발생했다면, except문 실행
    print('더보기 버튼 전부 클릭 완료')
```

```python
# 이 한 셀만 실행해도 한솥 도시락 페이지의 모든 데이터가 수집될 수 있도록
# 코드 조각을 끼워맞춰 보자

# 라이브러리 불러오기
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import requests as req
from bs4 import BeautifulSoup as bs

### 한솥도시락 전체 메뉴 및 가격 데이터 크롤링
url = 'https://www.hsd.co.kr/menu/menu_list'
s = Service(ChromeDriverManager().install())

driver = wb.Chrome(service = s)
driver.get(url)

# bs 객체화
soup = bs(driver.page_source, 'lxml')
soup

# 메뉴 이름
menu = soup.select('h4.h.fz_03')
menu

# 가격
price = soup.select('div.item-price>strong')
price

# 반복문 써서 이름과 가격 모두 수집하기
menulst = []
pricelst = []
for i in range(len(menu)):
    menulst.append(menu[i].text)
    pricelst.append(price[i].text)
print(menulst)
print(pricelst)

# 더보기 버튼 누르기 세팅
btn_more = driver.find_elements(By.CLASS_NAME, 'c_05')
btn_more = btn_more[0]
btn_more.click()

import time # time.sleep(몇 초?) => 몇 초만큼 기다렸다가 실행

try:
    # 에러가 나기 전까지 실행해야 할 실행문장 작성
    for i in range(300):
        time.sleep(2) # 2초의 쉬는 시간이 부여됨 (페이지가 무거울 때 안정성 증가)
        btn_more = driver.find_elements(By.CLASS_NAME, 'c_05')
        btn_more = btn_more[0]
        btn_more.click()        
except:
    # try문을 실행하다가 오류가 발생했다면, except문 실행
    print('더보기 버튼 전부 클릭 완료')
```
